### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# CLEAN DATA
# -----------------------------
df_db = df[['DatabaseWantToWorkWith']].dropna()

# Split multiple databases into rows
df_db = df_db.assign(
    Database=df_db['DatabaseWantToWorkWith'].str.split(';')
).explode('Database')

# -----------------------------
# COUNT FREQUENCY
# -----------------------------
top_db = df_db['Database'].value_counts().head(5)

# -----------------------------
# PIE CHART
# -----------------------------
plt.figure(figsize=(6,6))

plt.pie(
    top_db,
    labels=top_db.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Top 5 Databases Respondents Want to Work With")

plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# CLEAN DATA
# -----------------------------
df_dev = df[['DevType']].dropna()

# Split multiple roles into rows
df_dev = df_dev.assign(
    DevRole=df_dev['DevType'].str.split(';')
).explode('DevRole')

# -----------------------------
# COUNT TOP 5 ROLES
# -----------------------------
top_roles = df_dev['DevRole'].value_counts().head(5)

# -----------------------------
# PIE CHART
# -----------------------------
plt.figure(figsize=(7,7))

plt.pie(
    top_roles,
    labels=top_roles.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Top 5 Developer Roles (DevType Distribution)")

plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Find correct column automatically
ai_col = [col for col in df.columns if "AI" in col and "Tool" in col]

print("Detected column:", ai_col)

# Use first match
col = ai_col[0]

df_ai = df[[col]].dropna()

df_ai = df_ai.assign(
    Tool=df_ai[col].str.split(';')
).explode('Tool')

top_ai = df_ai['Tool'].value_counts().head(5)

plt.figure(figsize=(7,7))

plt.pie(
    top_ai,
    labels=top_ai.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Top 5 AI Development Tools")

plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# SAFE CLEANING
# -----------------------------
df_fw = df[['WebframeWantToWorkWith']].dropna()

# Split frameworks into rows
df_fw = df_fw.assign(
    Framework=df_fw['WebframeWantToWorkWith'].str.split(';')
).explode('Framework')

# Remove extra spaces (important improvement)
df_fw['Framework'] = df_fw['Framework'].str.strip()

# -----------------------------
# TOP 5 FRAMEWORKS
# -----------------------------
top_fw = df_fw['Framework'].value_counts().head(5)

# -----------------------------
# PIE CHART
# -----------------------------
plt.figure(figsize=(7,7))

plt.pie(
    top_fw.values,
    labels=top_fw.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Top 5 Preferred Web Frameworks")

plt.show()
```
